```python
def testColumn(name):
    df = energy_df[name]
    clean_data = pd.to_numeric(df, errors='coerce').dropna()
    
    q1 = clean_data.quantile(0.25)
    q3 = clean_data.quantile(0.75)

    iqr = q3 - q1

    t_low = q1 - (1.5 * iqr)
    t_high = q3 + (1.5 * iqr)
    
    outliers = [];
    for x in clean_data:
        if x < t_low or x > t_high:
            outliers.append(x)
    sorted(outliers, reverse=True)
    
    screen_range = (75, 775)
    curr_range = (clean_data.min(), clean_data.max())
    scale = lambda val : ((val - curr_range[0]) / (curr_range[1]-curr_range[0])) * (screen_range[1]-screen_range[0]) + screen_range[0]

    return { 'data': clean_data, 'summary': { 
                'count': len(clean_data),
                'range': curr_range, 
                'median': clean_data.median(),
                'q1': q1, 
                'q3': q3, 
                'iqr': iqr, 
                't_low': t_low, 
                't_high': t_high
            },
               'outliers': outliers, 'scale': scale }
    
# print(scale(10354, (25.0, 87206.0), ))
    
```

```python
electric = testColumn('AVERAGE BUILDING AGE')
```

```python
for k, v in electric['summary'].items():
    print(f'{k}:\t{v}')
```
